```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('user.csv')

# Display the first few rows of the dataframe
print("First few rows of the dataset:")
print(df.head())

# Display basic information about the dataset
print("\nDataset info:")
print(df.info())
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot 1: Box plot of age distribution by gender
sns.boxplot(x='gender', y='age', data=df, ax=ax1)
ax1.set_title('Age Distribution by Gender')

# Plot 2: Box plot of age distribution by occupation
sns.boxplot(x='occupation', y='age', data=df, ax=ax2)
ax2.set_title('Age Distribution by Occupation')
plt.setp(ax2.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
